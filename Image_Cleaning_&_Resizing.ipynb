```python
import cv2
import numpy as np
import os

# Create standard directories for testing
os.makedirs("raw_images", exist_ok=True)
os.makedirs("cleaned_images", exist_ok=True)

# Generate 3 dummy "messy" images
for i in range(3):
    # Create a dark, low-contrast gray image
    dummy_img = np.random.randint(20, 80, (500, 600, 3), dtype=np.uint8)
    cv2.imwrite(f"raw_images/img_{i}.jpg", dummy_img)

print("Generated dummy images in 'raw_images' folder.")
```

```python
def clean_and_resize(image_path, target_size=(224, 224)):
    # Read the image in grayscale for simple contrast adjustment
    img = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
    if img is None:
        return None

    # Resize the image smoothly
    resized = cv2.resize(img, target_size, interpolation=cv2.INTER_AREA)

    # Apply Adaptive Histogram Equalization (CLAHE) to fix dark/poor lighting
    clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8, 8))
    equalized = clahe.apply(resized)

    return equalized
```

```python
raw_folder = "raw_images"
output_folder = "cleaned_images"

for file_name in os.listdir(raw_folder):
    if file_name.endswith((".jpg", ".jpeg", ".png")):
        input_path = os.path.join(raw_folder, file_name)
        processed_img = clean_and_resize(input_path)

        if processed_img is not None:
            output_path = os.path.join(output_folder, f"clean_{file_name}")
            cv2.imwrite(output_path, processed_img)

print("Pipeline successful! Check the 'cleaned_images' folder.")
```
